```python
from transformers import pipeline
import matplotlib.pyplot as plt
sentiment_pipeline = pipeline("sentiment-analysis")
reviews = [
    "I absolutely love this product!",
    "It's the worst purchase I've ever made.",
    "Meh, it’s okay. Nothing special.",
    "Fantastic quality and quick shipping.",
    "Terrible customer service experience.",
    "Highly recommended for anyone!",
    "Would not buy this again.",
    "Pretty decent for the price.",
    "Extremely disappointed with the results.",
    "Great value and works as expected."
]
results = sentiment_pipeline(reviews)
for i, res in enumerate(results):
    print(f"{i+1}. {reviews[i]}")
    print(f"   Sentiment: {res['label']} | Confidence: {round(res['score'], 4)}")
labels = [r['label'] for r in results]
label_counts = {label: labels.count(label) for label in set(labels)}
colors = ['#4CAF50', '#F44336', '#FF9800']
plt.figure(figsize=(6, 6))
plt.pie(label_counts.values(), labels=label_counts.keys(), autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Sentiment Analysis Distribution")
plt.axis('equal')
plt.show()
```
